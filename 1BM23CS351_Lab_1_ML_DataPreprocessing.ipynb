```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

diabetes_df = pd.read_csv("Dataset of Diabetes.csv")


# Display basic info
print("Dataset Info:")
diabetes_df.info()
```

```python
numeric_cols = diabetes_df.select_dtypes(include=np.number)
numeric_cols.fillna(numeric_cols.mean(), inplace=True)
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1

numeric_cols = numeric_cols[
    ~((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)
]
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
diabetes_minmax = minmax_scaler.fit_transform(numeric_cols)
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
diabetes_standardized = standard_scaler.fit_transform(numeric_cols)
```

```python
# ADULT INCOME DATASET PREPROCESSING

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Load dataset

adult_df = pd.read_csv("adult.csv.zip")

print("Dataset Info:")
adult_df.info()

print("\nFirst 5 rows:")
print(adult_df.head())
```

```python
# Replace '?' with NaN
adult_df.replace('?', np.nan, inplace=True)

# Fill missing categorical values with mode
for col in adult_df.select_dtypes(include='object').columns:
    adult_df[col].fillna(adult_df[col].mode()[0], inplace=True)
```

```python
le = LabelEncoder()

for col in adult_df.select_dtypes(include='object').columns:
    adult_df[col] = le.fit_transform(adult_df[col])
```

```python
# Select only numerical columns
numeric_cols = adult_df.select_dtypes(include=np.number)

Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1

numeric_cols = numeric_cols[
    ~((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)
]
```

```python
minmax_scaler = MinMaxScaler()
adult_minmax = minmax_scaler.fit_transform(numeric_cols)
```

```python
standard_scaler = StandardScaler()
adult_standardized = standard_scaler.fit_transform(numeric_cols)
```
